```python
 # Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df.head()
```

```python
 # Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
clean_df = application_df.drop(columns=['EIN'])
clean_df.head()
```

```python
 # Determine the number of unique values in each column.
clean_df.nunique()
```

```python
# Look at NAME counts for binning
name_counts = application_df['NAME'].value_counts()
name_counts
```

```python
# Choose a cutoff value and create a list of names_replace to be replaced
# use the variable name names_replace
names_replace = list(name_counts[name_counts<5].index)
names_replace
```

```python
# Replace in dataframe
for name in names_replace:
    clean_df['NAME'] = clean_df['NAME'].replace(name,"Other")

# Check to make sure binning was successful
clean_df['NAME'].value_counts()
```

```python
# Look at APPLICATION TYPE counts for binning
name_counts2 = clean_df['APPLICATION_TYPE'].value_counts()
name_counts2
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(name_counts2[name_counts2<500].index)
application_types_to_replace

# Replace in dataframe
for app in application_types_to_replace:
    clean_df['APPLICATION_TYPE'] = clean_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
clean_df['APPLICATION_TYPE'].value_counts()

```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts 
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list (classification_counts [classification_counts <100].index)
classifications_to_replace

# Replace in dataframe
for i in classifications_to_replace:
    clean_df['CLASSIFICATION'] = clean_df['CLASSIFICATION'].replace(i,"Other")
    
# Check to make sure binning was successful
clean_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
numeric_df = pd.get_dummies(clean_df)
numeric_df.head()
```

```python
 # Split our preprocessed data into our features and target arrays
y = numeric_df['IS_SUCCESSFUL'].copy()
X = numeric_df.drop(columns = 'IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
 # Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=7, activation="relu", input_dim=X_train.shape[1]))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=14, activation="relu"))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the Sequential model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn_model.save('AlphabetSoupCharity_Optimization.h5')
```
